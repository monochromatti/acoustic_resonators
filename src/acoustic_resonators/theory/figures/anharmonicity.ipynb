```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x = np.linspace(0.1, 2, 1000)


# Lennard-Jones potential
V_LD = 4 * ((1 / x) ** 12 - (1 / x) ** 6)

# Harmonic potential
V_harmonic = 30 * (x - 2 ** (1 / 6)) ** 2 + np.min(V_LD)

fig, ax = plt.subplots(figsize=(6.8, 3.4))
ax.plot(x, V_harmonic, c=".5")
ax.plot(x, V_LD, c="k")
ax.axvline(x=0, c=".2", lw=0.5)
ax.axhline(y=0, c=".2", lw=0.5)

ax.set(xticks=[], yticks=[], ylim=(-1, 1))

ax.spines[:].set_visible(False)

fig.savefig("harmonic_vs_lennard_jones.svg", transparent=True)
```
